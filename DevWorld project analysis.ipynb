```python
Age_Grouping=dx['Age Group']
Total=dx['Sum']
```

```python
plt.figure(figsize=(7,5))
plt.xticks(rotation=0)

plt.bar(Age_Grouping,Total,color='red')
plt.xlabel('sum')
plt.ylabel('gruping')
plt.title('Customers Age Range')
plt.show
```
